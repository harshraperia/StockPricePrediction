# ***COMPARISON OF TOTAL TRADED OF STOCKS***

**Q1 WHICH AMONG THE THREE OF THE STOCKS HAS BEEN TRADED THE MOST IN A DAY?**

Here we are using MATPLOTLIP.PYPLOT to plot a graph for comparing the total traded stocks

```python
drrdy['Total traded'].plot(label = 'DRREDDY', figsize= (16,8))
hcl['Total traded'].plot(label = 'HCL TECH', figsize=(16,8))
tcsdf['Total traded'].plot(label = 'TCS', figsize=(16,8))
plt.title('COMPARISON OF TOTAL TRADED')
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.legend()

maxdr = drrdy['Total traded'].max()
print('Maximum of DRREDDY total traded stocks in a day', maxdr)
maxhcl = hcl['Total traded'].max()
print('Maximum of HCLTECH total traded stocks in a day', maxhcl)
maxtcs = tcsdf['Total traded'].max()
print('Maximum of TCS total traded stocks in a day', maxtcs)
maxlst = [maxdr, maxhcl, maxtcs]
mx = (maxdr if (maxdr > maxhcl and maxdr > maxtcs) else
     (maxhcl if (maxhcl > maxdr and maxhcl > maxtcs) else maxtcs))
if mx == maxdr:
  print('DRREDDY stocks have been traded the most in a day')
if mx == maxhcl:
  print("HCLTECH stocks have been traded the most in a day")
if mx == maxtcs:
  print("TCS stocks have been traded the most in a day")
```

# ***PLOT FOR THE MOVING AVERAGES***

Here we have used tow function of pandas rolling and mean, where rolling defines the window for which the mean is to be calculated.

```python

drrdy['MA1']=""
drrdy['MA2']=""
hcl['MA1']=""
hcl['MA2']=""
tcsdf['MA1']=""
tcsdf['MA2']=""



drrdy['MA1'] = drrdy['Open'].rolling(50).mean()
drrdy['MA2'] = drrdy['Open'].rolling(200).mean()
drrdy[['Open','MA1','MA2']].plot(label = 'DRREDDY', figsize=(16,10));
plt.title('DRREDDY MOVING AVERAGES')

hcl['MA1'] = hcl['Open'].rolling(50).mean()
hcl['MA2'] = hcl['Open'].rolling(200).mean()
hcl[['Open','MA1','MA2']].plot(label = 'HCL TECH', figsize=(16,10));
plt.title('HCL TECH MOVING AVERAGES')

tcsdf['MA1'] = tcsdf['Open'].rolling(50).mean()
tcsdf['MA2'] = tcsdf['Open'].rolling(200).mean()
tcsdf[['Open','MA1','MA2']].plot(label = 'TCS', figsize=(16,10));
plt.title('TCS MOVING AVERAGES')
```

# ***VOLATILITY***

**Q3 WHICH AMONG THE THREE STOCKS IS THE MOST VOLATILE STOCK?**

Here we are using the hist function of pyplot which helps us in creating a histogram for the values 

Volatility is a statistical measure of the dispersion of returns for a given security or market index.

```python
drrdy['returns']=""
hcl['returns']=""
tcsdf['returns']=""
drrdy['returns'] = (drrdy['Close']/drrdy['Close'].shift(1)) -1
hcl['returns'] = (hcl['Close']/hcl['Close'].shift(1))-1
tcsdf['returns'] = (tcsdf['Close']/tcsdf['Close'].shift(1)) - 1
drrdy['returns'].hist(bins = 100, label = 'DRREDDY', alpha = 0.5, figsize = (15,7))
hcl['returns'].hist(bins = 100, label = 'HCLTECH', alpha = 0.5)
tcsdf['returns'].hist(bins = 100, label = 'TCS', alpha = 0.5)
plt.legend()
#volatility
```

**ANS:** From the graph plotted above it is clearly visible that DRREDDY stocks are the most volatile stocks.

**Q4 WHICH AMONG THE THREE STOCKS HAS GIVEN THE BEST RETURNS TILL NOW ?**

```python
m1 = drrdy['returns'].mean()
m2 = hcl['returns'].mean()
m3 = tcsdf['returns'].mean()
mx = (m1 if (m1 > m2 and m1 > m3) else
     (m2 if (m2 > m1 and m2 > m3) else m3))
if mx == m1:
  print('DRREDDY is the stock which has given best returns')
if mx == m2:
  print("HCLTECH is the stock which has given best returns")
if mx == m3:
  print("TCS is the stock which has given best returns")
```
